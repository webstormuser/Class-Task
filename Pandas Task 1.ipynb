```python
#Explore over an internet that how much calories burn is 
#required for a healthy person and find out how many healthy person we have in our dataset-->for healthy life i considered 2000 calories must be burnt
def caloryCheck(a):
    if a>2000:
        return 1
    else :
        return 0
```

```python
df['HealthyLife']=df['Calories'].apply(caloryCheck)
df
#finding how many of dataset are burning calories to get Healthy life 
df[['Id','HealthyLife']].groupby('Id')['HealthyLife'].sum().sort_values(ascending=False) # these are ids of ttoal healthy person
```

```python
#how many person are not a regular person with respect to activity try to find out those
#We can consider 
#df.drop(columns='LoggDay',inplace=True)
healthy_df=df.groupby('Id')[['HealthyLife','Calories']].sum()
healthy_df.head()
```

```python
#how many person are not a regular person with respect to activity try to find out those
#for this we have to check how many people not even logged 
healthy_df['LoggDays']=df.groupby('Id')['ActivityDate'].count().sort_values(ascending=False)
healthy_df
```

```python
healthy_df['AvgCalory']=healthy_df['Calories']/healthy_df['LoggDays']
healthy_df
```

```python
healthy_df['AvgHealthIndex']=healthy_df['HealthyLife']/healthy_df['LoggDays']
healthy_df
```

```python
healthy_df.sort_values(by=['AvgCalory','LoggDays'],ascending=False).head()
```

```python
#who is the thired most active person in this dataset find out those in pandas and in sql both
# i am considering 50% value of Avg Healthindex to check person is healthy or unhealthy
def checkHealthiness(a,b):
    if ((a > 0.66) or(b > 14442)):
        return "Healthy"
    else :
        return "Unhealthy"
```

```python
checkHealthiness(0.89,1000)
```

```python
healthy_df['Health_Status']=healthy_df.apply(lambda healthy_df:checkHealthiness(healthy_df['AvgHealthIndex'],healthy_df['Calories']),axis=1)
```
